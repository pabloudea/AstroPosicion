```python
import numpy as np
import datetime
import pandas as pd

cos, sin, tan = np.cos, np.sin, np.tan
arcsin, arccos, arctan = np.arcsin, np.arccos, np.arctan
pi = np.pi
torad = pi/180
ToSideral = 1.00273790935079

def to_decimal(theta):
        '''Ingresado un ángulo en forma de cadena de caracteres (gra°min′sec″Hem), lo convierte a su forma decimal en °.
        Retorna el ángulo grados [°] de forma decimal y el hemisferio Hem
        '''
        grado, minu, seg, hem=int(theta.split('°')[0]), int(theta.split('°')[1].split('′')[0]), int(theta.split('°')[1].split('′')[1].split('″')[0]), theta.split('°')[1].split('′')[1].split('″')[1]
        grados = [grado + minu/60 +seg/3600 if (hem == 'N' or hem =='E') else -(grado + minu/60 +seg/3600) ][0]
        return (grados, hem)

        
def coord(pos):
    '''Revisa el tipo de formato en el que se ingresaron las coordenadas.
    Si están en formato decimal, se toman como fueron ingresadas De lo contrario, convierte las coordenadas ingresadas 
    a forma decimal
    '''
    lat=lon=0.
    if (type(pos[0])== float or type(pos[0])==np.float64):
        lat = pos[0]
        lon = pos[1]
    else:
        lat = to_decimal(pos[0])[0] #Transforma la latitud del punto P_0 a forma decimal
        lon = to_decimal(pos[1])[0] #Transforma la longitud del punto P_0 a forma decimal
    return lat,lon
```

```python
def JD(date):
    '''Calcula el día juliano para la fecha determinada
    date: Tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    Ejemplo:
    date = (Year,Month,Day,hour,minute,second)= (2000,06,21,12,30,15.3) corresponde a las 12:30:15.3 horas del 21/Junio/2000; 
    Retorna un float JDay con una precisión de 9 dígitos
    '''
    Yp,Mp,Dp,hp,mp,sp = date
    D_ = Dp+(hp+mp/60+sp/3600)/24
    if (Mp==1 or Mp==2):
        Yp,Mp = Yp-1,Mp+12
    A = int(Yp/100)
    B = 2-A+int(A/4)
    JDay = int(365.25*(Yp+4716))+int(30.6001*(Mp+1))+D_+B-1524.5
    return round(JDay,9)

def DateFromJD(jd):
    '''Devuelve una fecha tipo calendario (YYYY,MM,DD) con horas(hh), minutos(mm) y segundos(ss)
    jd: Julian Day (float) siempre positivo
    Retorna una tupla con 6 elementos: DateFromJD(x)->(Y,M,D,h,m,s)
    '''
    F,Z=np.modf(jd+.5) #F es la parte decimal de x y Z la parte entera i.e np.modf(1.68)->F=0.68 y Z=1
    if Z<2299161:
        A=Z
    else:
        alpha = int((Z-1867216.25)/36524.25)
        A = Z + 1 + alpha - int(alpha/4)
        B = A + 1524
        C = int((B-122.1)/365.25)
        D = int(365.25*C)
        E = int((B-D)/30.6001)
        
    day_ = round(B-D-int(30.6001*E)+F,8)
    day = int(day_)
    hour_,hour = (day_-day)*24,int((day_-day)*24)
    minu_,minu = (hour_-hour)*60,int((hour_-hour)*60)
    sec = round((minu_ - minu)*60,2)
    #Si sec=60, sec=0 y sumar un minuto
    if sec==60:
        sec=0
        minu= minu+1
    month = [E-1 if E<14 else E-13][0]
    year = [C-4715 if (month==1 or month==2) else C-4716][0]
    return year,month,day,hour,minu,sec

def interval_time(date1,date2):
    '''Calcula el tiempo transcurrido entre dos fechas date1 y date2 dadas. Para esto, las convierte a sus
    respectivos días julianos y da el resultado en días.
    date: Tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    '''
    jd1 = JD(date1)
    jd2 = JD(date2)
    result = jd1-jd2
    return result

def weekday(date):
    '''Dada una fecha, retorna un string con el día de la semana cayó. 
    date: Tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    '''
    if type(date)!=list:
        date = list(date)
    date[-3]=date[-2]=date[-1]=0
    days = {1:'Lunes',2:'Martes',3:'Miércoles',4:'Jueves',5:'Viernes',6:'Sábado',0:'Domingo'}
    reminder = (JD(date)+1.5)%7
    return days[reminder]

def yearday(date):
    '''Calcula que día del año fue una fecha dada, siendo 1 el 1 de enero y 365/366 el 31 de diciembre. Devuelve un float
    date: Tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    '''
    #date = (Y,M,D,h,m,s)
    K = [1 if date[0]%4==0 else 2][0]
    N = int(275*date[1]/9) - K*int((date[1]+9)/12) + (date[2]+date[3]/24+date[4]/1440) - 30
    #Devuelve el número del día en el año
    return N

def date_yearday(Nday,year):
    K = [1 if year%4==0 else 2][0]
    M = [1 if N<32 else int( 9*(K+N)/275 + 0.98 )][0]
    D = N - int(275*M/9)+K*int((M+9)/12)+30
    h = (D-int(D))*24
    m = (h-int(h))*60
    s = (m-int(m))*60
    date = [year,M,D,h,m,s]
    return date

def GST0(date):
    '''ángulo horario en horas que tiene el punto vernal a las 0h de TU para un observador situado exactamente en
    el meridiano de Greenwich) para una fecha determinada
    date: Tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    '''
    T = (JD(date)-2451545.0)/36525
    res = (6+41/60+50.54841/3600) + (2400+3/60+4.81286/3600)*T + 0.09310*(T**2)
    
    return round(res-(res//24)*24,9)

def LST(date,pos):
    Y,M,D,h,m,s = date
    lat,lon = coord(pos) #Latitude and longitude in degrees
    timezone = round(lon/15) #Time zone respect UTC. i.e -5 is -5 (UTC)
    UT = h+m/60+s/300+timezone
    GST = (GST0(date) + UT*ToSideral) - (GST0(date) + (UT*ToSideral)//24)*24
    lst = GST + timezone
    return lst
```

```python
def Sun_Pos(date,pos):
    '''pos: Coordenadas de posición del observador sobre la Tierra. pos=[latitude,longitude]. Si se ingresa un valor numérico
    se tomará este valor como la latitud para la cual se desea calcular la trayectoria del Sol ubicado desde el primer meridiano.
    Las coordenadas pueden ser ingresados en forma cadena de caracteres o en forma decimal como un arreglo de Python o NumPy
    date: Una tupla, lista o arreglo de 6 componentes: año(Y),Mes(M),Día(D),hora(h),minutos(m) y segundos(s)
    Ejemplo:
    pos = ['35°33′12″N', '139°46′52″E']; pos = np.array([25.2632, -80.97]); pos=6.16
    date = (Year,Month,Day,hour,minute,second)= (2000,06,21,12,30,15.3) corresponde a las 12:30:15.3 horas del 21/Junio/2000; 
    Retorna el Azimuth y el ángulo horario del Sol en la fecha date para un observador ubicado en las coordenadas pos
    i.e Sun_Pos(date,pos)=[Az,ha]
    '''
    if type(pos)!=list:
        pos=list(pos)
        #Si se quiere calcular el angulo horario y el azimuth de salida del Sol para una fecha y una latitud determinada, 
        #sin tener en cuenta su longitud, se calcula para el Primer meridiano, es decir, longitud=0
        if len(pos)==1:
            pos.append(0.) 
    
    Y,M,D,h,m,s = date
    lat,lon = coord(pos) #Latitude and longitude in degrees
    timezone = round(lon/15) #Time zone respect UTC. i.e -5 is -5 (UTC)
    
    gamma = 2*pi/365*(yearday(date)-1+(h-12)/24) #in radians
    #Declinación del Sol para una fecha dada
    decl = 0.006918 - 0.399912*cos(gamma) + 0.070257*sin(gamma) - 0.006758*cos(2*gamma) + 0.000907*sin(2*gamma) - 0.002697*cos(3*gamma) + 0.00148*sin(3*gamma) #in radians
    #Ecuación del tiempo para ese momento (diferencia entre el tiempo solar aparente y el tiempo solar medio)
    eqtime = 229.18*(0.000075 + 0.001868*cos(gamma) - 0.032077*sin(gamma) - 0.014615*cos(2*gamma) - 0.040849*sin(2*gamma) )
    
    #Ángulo horario del Sol al momento de su salida en radianes
    ha_sunrise = arccos(cos(90.833*torad)/(cos(lat*torad)*cos(decl)) - tan(lat*torad)*tan(decl))
    #Ángulo cenital solar en radianes al momento de salir(complemento de altura solar en coordenadas horizontales i.e: pi/2-h)
    SZA_sunrise = arccos( sin(lat*torad)*sin(decl) + cos(lat*torad)*cos(decl)*cos(ha_sunrise) )
    #Azimuth del Sol al momento de salir en radianes
    Az_sunrise = arccos( ( sin(decl)-sin(lat*torad)*cos(SZA_sunrise) )/( cos(lat*torad)*sin(SZA_sunrise) ) ) + pi 
    #Hora de salida (UTC) del Sol en minutos para ese día
    #Ej: 6h00m00s= 6h*60m/h =360minutos; 3h10m30s=3h*60m/h+10+30s/60s/m=190.5minutos, etc
    sunrise = 720-4*(lon+ha_sunrise/torad)-eqtime
    rise_dict = {'HA_sunrise':ha_sunrise,'Az_sunrise':Az_sunrise,'time_sunrise':sunrise/60+timezone}
    
    #Ángulo horario del Sol al momento de su ocaso en radianes
    ha_sunset = -1.*ha_sunrise
    #Ángulo cenital solar en radianes al momento de ponerse(complemento de altura solar en coordenadas horizontales i.e: pi/2-h)
    SZA_sunset = arccos( sin(lat*torad)*sin(decl) + cos(lat*torad)*cos(decl)*cos(ha_sunset) )
    #Azimuth del Sol al momento de ponerse en radianes
    Az_sunset = arccos( ( sin(decl)-sin(lat*torad)*cos(SZA_sunset) )/( cos(lat*torad)*sin(SZA_sunset) ) ) + pi
    #Hora de puesta (UTC) del Sol en minutos para ese día
    #Ej: 18h00m00s= 18h*60m/h =1080minutos; 19h10m30s=19h*60m/h+10+30s/60s/m=1150.5minutos, etc
    sunset = 720-4*(lon+ha_sunset/torad)-eqtime
    set_dict = {'HA_sunset':ha_sunrise,'Az_sunset':Az_sunrise,'time_sunset':sunset/60+timezone}
    
    #Desfase horario respecto a la zona de tiempo
    time_offset = eqtime + 4*lon - 60*timezone
    #Tiempo solar verdadero (true solar time) en minutos
    tst = h*60 + m + s/60 + time_offset
    
    #Ángulo horario del Sol en la fecha y hora determinadas en date
    ha = ((tst / 4) - 180) #Solar Hour Angle in degrees
    #Ángulo cenital solar en radianes en la fecha y hora determinadas en date
    SZA = arccos( sin(lat*torad)*sin(decl) + cos(lat*torad)*cos(decl)*cos(ha*torad) )
    #Azimuth del Sol al momento en radianes
    Az = arccos( ( sin(decl)-sin(lat*torad)*cos(SZA) )/( cos(lat*torad)*sin(SZA) ) ) + pi
    date_dict = {'HA_date':ha,'Az_date':Az,'TrueSolarTime':tst/60+timezone}
    return rise_dict,set_dict,date_dict
```

```python
def Horizontal_a_Horaria(A, h, phi):
    '''Los ángulos deben ser ingresados en grados ° y se debe tener en cuenta que una latitud negativa corresponde a un
    observador ubicado en el hemisferio sur terrestre
    A=Azimut: 0<A<360; h=altura: -90<h<90, phi=latitud: -90<phi<90
    Devuelve un arreglo con la declinación dec y el ángulo horario H en horas [dec,H]
    Ejemplos:
    A1,h1,phi1=210+34/60, 35+43/60, 3+25/60
    dec1,H1 = Horizontal_a_Horaria(A1,h1,phi1)
    print('La declinación es: %.d°%.2f\' y el ángulo horario es: %.dh %.2fmin '
    %(int(dec1),(dec1-int(dec1))*60,int(H1),(H1-int(H1))*60 ) )
    A2,h2,phi2= 47+34/60, 67+45/60, -(17+36/60)
    dec2,H2 = Horizontal_a_Horaria(A2,h2,phi2)
    print('La declinación es: %.d°%.2f\' y el ángulo horario es: %.dh %.2fmin '
    %(int(dec2),(dec2-int(dec2))*60,int(H2),(H2-int(H2))*60 ) )
    '''
    A_,h_,phi_=A*torad,h*torad,phi*torad
    dec = arcsin(sin(phi_)*sin(h_) + cos(phi_)*cos(h_)*cos(A_))
    H = [arccos((sin(h_)-sin(dec)*sin(phi_))/(cos(dec)*cos(phi_)) ) if A>180 
         else 2*pi-arccos((sin(h_)-sin(dec)*sin(phi_))/(cos(dec)*cos(phi_)))][0]/(torad*15)
    return [dec/torad,H]

def Horaria_a_Horizontal(dec, H, phi):
    '''Los ángulos deben ser ingresados en grados ° (ejemplo:6.1432°) y se debe tener en cuenta que una latitud negativa 
    corresponde a un observador ubicado en el hemisferio sur terrestre
    El ángulo horario debe ser ingresado en horas y en formato decimal (ejemplo: 4.333, que representa 4.333 horas)
    H=Ángulo horario: 0<H<24h; dec=declinación: -90<dec<90, phi=latitud: -90<phi<90
    Devuelve un arreglo con la altura h y el azimut del astro para un observador ubicado en una latitud phi [h,A]
    Ejemplos:
    dec1, H1, phi1 = 34+14/60,5+35.3/60,1+9/60 #Para observador ubicado en Mocoa (phi=1°9')
    A1,h1=Horaria_a_Horizontal(dec1,H1,phi1)
    print('El acimut es: %.d°%.2f\' y la altura es: %.d° %.2f\' '%(int(A1),(A1-int(A1))*60,int(h1),(h1-int(h1))*60 ) )
    dec2, H2, phi2 = -(8+12/60),20+45.1/60,10+27/60 #Rigel(dec=8°12') para observador en Cartagena(phi=10°27') en cierto momento
    A2,h2=Horaria_a_Horizontal(dec2,H2,phi2)
    print('El acimut es: %.d°%.2f\' y la altura es: %.d° %.2f\' '%(int(A2),(A2-int(A2))*60,int(h2),(h2-int(h2))*60 ) )
    '''
    dec_,H_,phi_=dec*torad,(H*15)*torad,phi*torad
    h = arcsin(sin(phi_)*sin(dec_) + cos(phi_)*cos(dec_)*cos(H_))
    A = [arccos( (sin(dec_) - sin(h)*sin(phi_))/(cos(h)*cos(phi_)) ) if H>12
         else 2*pi-arccos((sin(dec_)-sin(h)*sin(phi_))/(cos(h)*cos(phi_)))][0]/torad
    return [h/torad,A]

def Horaria_a_Ecuatorial(dec,H,TSL=[list(datetime.datetime.now().timetuple()[:6]),[0,0]] #TSL=[date=[Y,M,D,h,m,s],pos=[lat,lon]]
                        ):
    if type(TSL)==list or type(TSL)==np.array:
        TSL=LST(TSL[0],TSL[1])
    ra = [TSL - H if TSL>H else 24+(TSL-H)][0]
    return [dec,ra]

def Ecuatorial_a_Horaria(dec,ra,TSL=[list(datetime.datetime.now().timetuple()[:6]),[0,0]] #TSL=[date=[Y,M,D,h,m,s],pos=[lat,lon]]
                        ):
    if type(TSL)==list or type(TSL)==np.array:
        TSL=LST(TSL[0],TSL[1])
    H = [TSL - ra if TSL>ra else 24+(TSL-ra)][0]
    return [dec,H]

def Ecuatorial_a_Ecliptica(ra,dec):
    obl = 23+26/60+11/3600
    ra = ra*15 - ((ra*15)//360)*360
    p = sin(dec*torad)*sin(obl*torad)+cos(dec*torad)*cos(obl*torad)*sin(ra*torad)
    q = cos(ra*torad)*cos(dec*torad)
    
    ecl_lat = arcsin(sin(dec*torad)*cos(obl*torad) - cos(dec*torad)*sin(obl*torad)*sin(ra*torad))/torad
    ecl_lon = [180+arctan(p/q)/torad if (p*q<0 and q<0) or p+q<0 else [360+arctan(p/q)/torad if (p*q<0 and q>0) else arctan(p/q)/torad ][0] ][0]
    return [ecl_lon,ecl_lat]

def Ecliptica_a_Ecuatorial(ecl_lon,ecl_lat):
    obl = (23+26/60)
    p = -sin(ecl_lat*torad)*sin(obl*torad)+cos(ecl_lat*torad)*cos(obl*torad)*sin(ecl_lon*torad)
    q = cos(ecl_lon*torad)*cos(ecl_lat*torad)
    dec = arcsin(sin(ecl_lat*torad)*cos(obl*torad) + cos(ecl_lat*torad)*sin(obl*torad)*sin(ecl_lon*torad))
    ra = [180+arctan(p/q)/torad if ((p*q<0 and q<0) or (p+q<0)) else [360+arctan(p/q)/torad if (p*q<0 and q>0) else arctan(p/q)/torad ][0] ][0]
    return [dec/torad,ra/15]

    
```

```python
def Eclipticas_a_Lunares(lon_ecl,lat_ecl,lam):
    lon_node = lam
    C = lon_node-lon_ecl
    i = 5.15 #°
    lun_dec = arcsin(sin(lat_ecl*torad)*cos(i*torad) + cos(lat_ecl*torad)*sin(i*torad)*sin(C*torad))/torad
    p = sin(ecl_lat*torad)*sin(i*torad)-cos(ecl_lat*torad)*cos(i*torad)*sin(C*torad)
    q = cos(ecl_lat*torad)*cos(C*torad)
    lun_lon = [180+arctan(p/q)/torad if ((p*q<0 and q<0) or (p+q<0)) else [360+arctan(p/q)/torad if (p*q<0 and q>0) else arctan(p/q)/torad ][0] ][0]
    return [lun_dec,lun_lon/torad]
def Lunares_a_Ecliptica(lun_dec,lun_lon):
    lon_node = lam
    #C = lon_node-lon_ecl
    i = 5.15 #°
    lat_ecl = arcsin(sin(lun_dec*torad)*cos(i*torad) + cos(lun_dec*torad)*sin(i*torad)*sin(lun_lon*torad))/torad
    p = sin(lun_dec*torad)*sin(i*torad)-cos(lun_dec*torad)*cos(i*torad)*sin(lun_lon*torad)
    q = cos(lun_dec*torad)*cos(lun_lon*torad)
    lon_ecl = [180+arctan(p/q)/torad if ((p*q<0 and q<0) or (p+q<0)) else [360+arctan(p/q)/torad if (p*q<0 and q>0) else arctan(p/q)/torad ][0] ][0]
    return [lat_ecl,lon_ecl/torad]

def lunar_long(deltaT, w_luna=13.176359, w_node=0.0529907 ): #w_luna y w_node están en unidades [°/dia]
    new_long = (w_node+w_luna)*deltaT - ((w_node+w_luna)*deltaT//360)*360
    return new_long
```
